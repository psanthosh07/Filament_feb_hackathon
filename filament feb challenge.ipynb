```python
import pandas as pd

# Load data from the CSV file
btc_data = pd.read_csv('BTC-USD.csv')

# Convert the 'Date' column to datetime format
btc_data['Date'] = pd.to_datetime(btc_data['Date'])

# Bitcoin halving dates
halving_dates = ['2016-07-09', '2020-05-11']

# Analyze data before and after each halving event
for halving_date in halving_dates:
    halving_date = pd.to_datetime(halving_date)

    # Segment the data before and after the halving date
    before_halving = btc_data[btc_data['Date'] < halving_date]['Close']
    after_halving = btc_data[btc_data['Date'] >= halving_date]['Close']

    # Calculate average closing price before and after the halving
    avg_price_before = before_halving.mean()
    avg_price_after = after_halving.mean()

    # Print analysis results
    print(f"Analysis for Bitcoin halving on {halving_date.strftime('%Y-%m-%d')}:")
    print(f"Average closing price before halving: {avg_price_before}")
    print(f"Average closing price after halving: {avg_price_after}")
    print(f"Percentage change: {((avg_price_after - avg_price_before) / avg_price_before) * 100:.2f}%\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('BTC-USD.csv')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' column as the index
df.set_index('Date', inplace=True)

# Define the Bitcoin halving dates
halving_dates = ['2016-07-09', '2020-05-11']

# Create empty lists to store average closing prices before and after each halving event
avg_before_halving = []
avg_after_halving = []

# Iterate through halving dates
for halving_date in halving_dates:
    # Filter data before and after the halving date
    before_halving = df[df.index < halving_date]['Close']
    after_halving = df[df.index >= halving_date]['Close']

    # Calculate average closing prices
    avg_before_halving.append(before_halving.mean())
    avg_after_halving.append(after_halving.mean())

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(halving_dates, avg_before_halving, marker='o', label='Before Halving', color='blue')
plt.plot(halving_dates, avg_after_halving, marker='o', label='After Halving', color='orange')
plt.title('Average Closing Prices Before and After Bitcoin Halving')
plt.xlabel('Halving Event')
plt.ylabel('Average Closing Price (USD)')
plt.xticks(halving_dates)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Ethereum data
eth_data = pd.read_csv('ETH.csv', sep=';')

# Convert 'timestamp' column to datetime
eth_data['timestamp'] = pd.to_datetime(eth_data['timestamp'])

# Set 'timestamp' column as the index
eth_data.set_index('timestamp', inplace=True)

# Plot relevant metrics over time
plt.figure(figsize=(12, 8))

# Plotting open, high, low, close prices
plt.plot(eth_data.index, eth_data['open'], label='Open', color='blue')
plt.plot(eth_data.index, eth_data['high'], label='High', color='green')
plt.plot(eth_data.index, eth_data['low'], label='Low', color='red')
plt.plot(eth_data.index, eth_data['close'], label='Close', color='orange')

plt.title('Ethereum Price Movement Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Plotting volume and market cap
plt.figure(figsize=(12, 8))

plt.plot(eth_data.index, eth_data['volume'], label='Volume', color='blue')
plt.plot(eth_data.index, eth_data['marketCap'], label='Market Cap', color='green')

plt.title('Ethereum Volume and Market Cap Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Bitcoin data
bitcoin_data = pd.read_csv('BTC-USD.csv')

# Load Ethereum data with the correct delimiter
ethereum_data = pd.read_csv('ETH.csv', sep=';')

# Convert 'Date' column to datetime format
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])
ethereum_data['timestamp'] = pd.to_datetime(ethereum_data['timestamp'])

# Set 'Date' column as the index for Bitcoin data
bitcoin_data.set_index('Date', inplace=True)

# Set 'timestamp' column as the index for Ethereum data
ethereum_data.set_index('timestamp', inplace=True)

# Plot Bitcoin and Ethereum prices over time
plt.figure(figsize=(12, 6))
plt.plot(bitcoin_data.index, bitcoin_data['Close'], label='Bitcoin', color='orange')
plt.plot(ethereum_data.index, ethereum_data['close'], label='Ethereum', color='blue')

plt.title('Bitcoin vs Ethereum Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
